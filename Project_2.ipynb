```python
from Bio import SeqIO
from Bio import AlignIO
from Bio import pairwise2
from Bio.Seq import Seq
from Bio.pairwise2 import format_alignment
```

```python
# taking sequence data and saving it to variables
```

```python
demersus = SeqIO.read("demersus_nadh1.txt", "fasta")
humboldti = SeqIO.read("humboldti_nadh1.txt", "fasta")
magellanicus = SeqIO.read("magellanicus_nadh1.txt", "fasta")
mendiculus =SeqIO.read("mendiculus_nadh1.txt", "fasta")
```

```python
# pairwise sequence alignment
```

```python
alignments = pairwise2.align.globalxx(demersus, humboldti)
for alignment in alignments:
    print(format_alignment(*alignment))
```

```python
alignments = pairwise2.align.globalxx(demersus, magellanicus)
for alignment in alignments:
    print(format_alignment(*alignment))
```

```python
alignments = pairwise2.align.globalxx(demersus, mendiculus)
for alignment in alignments:
    print(format_alignment(*alignment))
```
